```python
# Import
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import os
```

```python
# Set path
Path = 'C:\\Users\\cschw\\OneDrive\\Desktop\\MLB Project'
```

```python
#upload df
mlb = pd.read_csv(os.path.join(Path, 'Data Sets', 'Main Data Set', 'MLB One Tab 10-9 Update.csv'))
```

```python
#check head
mlb.head()
```

```python
print(mlb.columns)
```

```python
# Create subset with only 2023 and 2024
mlb_sub = mlb[mlb['Year'].isin([2023, 2024])]
```

```python
# Creating a subset with selected columns
mlb_sub = mlb_sub[['Year', 'W', 'L', 'R', 'Home Wins', 'Home Losses', 'TR', 'H', 'HR', 'SB', 'BB', 'SO', 'ERA', 'Attendance', 'Attend/G', 'Avg Game Time Min']]
```

```python
# Check head
mlb_sub.head()
```

```python
# Create a scatter plot of Attendance and Avg Game Time Min
sns.lmplot(x='Avg Game Time Min', y='Attendance', data=mlb_sub)

# Disable scientific notation on both axes
plt.ticklabel_format(style='plain', axis='both')

# Adjust the font size for the y-axis ticks
plt.tick_params(axis='y', labelsize=8) 
plt.tick_params(axis='x', labelsize=10)

# Add a title
plt.title("Relationship Between Game Length and Attendance 23-24", fontsize = 10)

# Adjust the layout to prevent the labels from getting cut off
plt.tight_layout()

# Save the plot
plt.savefig(f"{Path}/attendance_time_2023_2024.png")
```
