### Exporting Info as CSV

```python
import io

# Capture the output of mlbdf.info()
buffer = io.StringIO()
mlbdf.info(buf=buffer)
info_str = buffer.getvalue()

# Split the captured output into lines and then into lists
info_lines = info_str.splitlines()
info_list = [line.split() for line in info_lines]

# Convert the list into a DataFrame
info_df = pd.DataFrame(info_list)

# Save the DataFrame to a CSV file
info_df.to_csv(r'C:\Users\cschw\OneDrive\Desktop\MLB Project\Data Sets\MLB_Info.csv', index=False, header=False)
```
